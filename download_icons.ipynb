```python
from PIL import Image



def make_transparent_green(file_in, file_out):
    img = Image.open(file_in)
    img = img.convert("RGBA")

    datas = img.getdata()

    newData = []

    for items in datas:
        if items[3] == 0:
            newData.append((0, 255, 255, 255))
        else:
            newData.append(items)

    img.putdata(newData)
    img.save(file_out, "png")
    print("Successful")


for file in glob.glob("D:\\Studies\\Introduction-To-Data-Science\\DataScienceFinalProject\\game_images\\compressed\\*.png"):
    filename = file.split('\\')[-1]
    make_transparent_green(file, f'./game_images/png_green/{filename}')
```
